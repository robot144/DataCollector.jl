```julia
# Matroos database

struct Matroos
    base_url::String
    login::String
    password::String
end

# Constructor with a base_url, login and password
function Matroos(base_url::String, login::String, password::String)
    return Matroos(base_url, login, password)
    # save to a file named ~/.matroos_credentials if this doesn't exist
    cred_file = joinpath(homedir(), ".matroos_credentials")
    if !isfile(cred_file)
        open(cred_file, "w") do f
            write(f, "base_url=$base_url\nlogin=$login\npassword=$password\n")
        end
    else
        open(cred_file, "r") do f
            for line in eachline(f)
                if startswith(line, "base_url=")
                    base_url = strip(line[10:end])
                elseif startswith(line, "login=")
                    login = strip(line[7:end])
                elseif startswith(line, "password=")
                    password = strip(line[9:end])
                end
            end
        end
    end
    return Matroos(base_url, login, password)
end
```
